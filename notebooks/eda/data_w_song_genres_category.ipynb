```python
import pandas as pd
```

```python
df_w_songs_genres = pd.read_csv('../data_w_song_genres/data_w_song_genres.csv')
```

```python
df_category = df_w_songs_genres[['artists','name']].copy()
```

```python
def categorize_popularity(value):
    if value <= 25:
        return "FLOP"
    elif 25 < value <= 50:
        return "MID"
    elif 50 < value <= 80:
        return "MID-HIT"
    
    return "HIT"

def categorize_numeric_features(value):
    if value <= 0.5:
        return 'Low'
    else:
        return 'High'

def categorize_bool_features(value):
    if value == 1 or value == True:
        return 'Yes'
    else:
        return 'No'

def categorize_mode(value):
    if value == 1:
        return 'Major'
    else:
        return 'Minor'


def categorize_key(value):
    key_dict = {
        -1: "Unknown",
        0: "C",
        1: "C♯/D♭",
        2: "D",
        3: "D♯/E♭",
        4: "E",
        5: "F",
        6: "F♯/G♭",
        7: "G",
        8: "G♯/A♭",
        9: "A",
        10: "A♯/B♭",
        11: "B",
    }
    return key_dict[value]

def categorize_duration(value):
    duration_dict = {
        (0, 60000): "Very short",
        (60001, 120000): "Short",
        (120001, 300000): "Mid",
        (300001, 480000): "Large",
        (480000, float("inf")): "Very large"
    }
    
    for (low, high), category in duration_dict.items():
        if low <= value < high:
            return category

def categorize_tempo(value):
    tempo_dict = {
        (0, 20): "Sin Tempo",
        (20, 40): "Grave",
        (40, 60): "Lento",
        (60, 66): "Largo",
        (66, 70): "Larghetto",
        (70, 76): "Adagio",
        (76, 80): "Adagietto",
        (80, 108): "Andante",
        (108, 112): "Moderato",
        (112, 120): "Allegretto",
        (120, 168): "Allegro",
        (168, 176): "Vivace",
        (176, 200): "Presto",
        (200, float("inf")): "Prestissimo"
    }

    for (low, high), category in tempo_dict.items():
        if low <= value < high:
            return category
    
    return 

def categorize_year(value):
    decade = (value // 10) * 10
    if value < 2000:
        return f"{decade % 100}s"
    else:
        return f"{decade}s"
```

```python
numeric_columns = [col for col in df_w_songs_genres.columns if col not in ["popularity","artists","duration_ms","id","key","mode","name","release_date","year","imputed_date","genres"]]

df_w_songs_genres["release_date"] = pd.to_datetime(df_w_songs_genres["release_date"])
```

```python
for col in numeric_columns:
    df_category[col] = df_w_songs_genres[col].apply(categorize_numeric_features)

df_category['popularity'] = df_w_songs_genres['popularity'].apply(categorize_popularity).astype('category')

df_category['duration'] = df_w_songs_genres['duration_ms'].apply(categorize_duration).astype('category')

df_category["tempo"] = (
    df_w_songs_genres["tempo"].apply(categorize_tempo).astype("category")
)

df_category["key"] = df_w_songs_genres["key"].apply(categorize_key).astype("category")

df_category['explicit'] = df_w_songs_genres['explicit'].apply(categorize_bool_features).astype('category')

df_category["imputed_date"] = (
    df_w_songs_genres["imputed_date"].apply(categorize_bool_features).astype("category")
)

df_category["mode"] = (
    df_w_songs_genres["mode"].apply(categorize_mode).astype("category")
)

df_category["month"] = (
    df_w_songs_genres["release_date"].dt.month_name().astype("category")
)

df_category['year'] = df_w_songs_genres['year'].apply(categorize_year).astype('category')
```
