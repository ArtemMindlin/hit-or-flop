```python
import pandas as pd
```

```python
df = pd.read_csv('data_formatted.csv')
df
```

```python
# Agrupar por artista y contar las canciones
songs_per_artist = df.groupby("artists").size().reset_index(name="song_count")

# Mostrar los primeros resultados
songs_per_artist.sort_values(by='song_count', ascending=False)
```
